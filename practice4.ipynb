```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.ebi.ac.uk/chembl/g/#browse/compounds','html.parser')
response = requests.get(url)
response
```

```python
data = response.text
data
```

```python
soup = BeautifulSoup(data,'lxml')
soup
```

```python
tags = soup.table
tags
```

```python
tag = soup.find_all('div')
tag
```

```python
a = []
for i in tag:
  s = i.get_text()
  a.append(s)

a
```

```python
final = a[1:]
final
```

```python
for i in final:
    f = i.split('\n')
    print(f)
```

```python
final_data = []
for i in final:
  f = i.split('\n')[1:9]
  final_data.append(f)

final_data
```

```python
output_data = final_data
output_data
```

```python
import csv
f = open('chembl.csv','w')
write = csv.writer(f)
for i in output_data:
  write.writerow(i)

f.close()
```
